# The Programming Brain

We will build an agent that acts like a Programmer: it writes and executes its own Python code to solve problems. This is perfect for FinTech because you don't need to pre-define tools for every possible financial metric (RSI, Bollinger Bands, Volatility)—the agent just writes the formula using pandas.

### Setup and API Configuration
This cell handles the initial setup. It installs the necessary `google-generativeai` library, imports it, and then configures the Gemini API using your API key. The API key is securely retrieved from Colab's user data secrets.

```python
# Setup (Use the standard library)
# Get your API key from https://aistudio.google.com/
!pip install -q -U google-generativeai # Ensure the correct library is installed
import google.generativeai as genai # Import the correct module to override previous
from google.colab import userdata
import os

os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
```

We will use `smolagents` and `litellm`. `smolagents` is the framework for building agents, and `litellm` is used to bridge non-Hugging Face models (like Gemini) to the Hugging Face ecosystem.

We will install the required libraries: `smolagents`, `litellm`, `yfinance`, `duckduckgo-search`, and `matplotlib`. It then imports `CodeAgent`, `DuckDuckGoSearchTool`, and `LiteLLMModel` from `smolagents`.

```python
# Install smolagents and dependencies
!pip install -q smolagents litellm yfinance duckduckgo-search matplotlib

from google.colab import userdata
from smolagents import CodeAgent, DuckDuckGoSearchTool, LiteLLMModel
```

#Initialize the Code Model
This Python code block configures the model. It initializes `LiteLLMModel` with `gemini/gemini-2.5-flash` and provides the API key. This connects the agent framework to the Gemini model.

```python
# Configure the Model
# smolagents uses 'LiteLLM' to connect to non-HuggingFace models easily.

model = LiteLLMModel(
    model_id="gemini/gemini-2.5-flash",
    api_key=os.environ["GOOGLE_API_KEY"]
)

print("✅ Model connected via LiteLLM")
```

This code block installs the `ddgs` package, which is a dependency for the `DuckDuckGoSearchTool` used by the agent. This ensures that the agent can perform web searches when needed.

```python
!pip install -q ddgs
```

This Python code block creates the `CodeAgent`. It's initialized with `DuckDuckGoSearchTool` for web searching and `LiteLLMModel` as its underlying model. Crucially, it's also given a list of `additional_authorized_imports` like `yfinance`, `pandas`, `numpy`, and `matplotlib.pyplot`, allowing the agent to use these libraries directly in its generated Python code.

```python
# Create the Code Agent
# We allow the agent to import specific libraries to do its job.

agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()], # We give it search for news
    model=model,
    max_steps=3,
    additional_authorized_imports=["yfinance", "pandas", "numpy", "matplotlib.pyplot"]
)

print("✅ Agent created. It can write and execute Python code.")
```

#Task 1 - The Calculation (No specific tool needed)
The goal here is to demonstrate the agent's ability to create its own logic for calculations, even for metrics like 'Volatility' that don't have a specific pre-built tool.


This Python code block defines and runs the first task. The agent is prompted to: 1. Get historical closing prices for 'NVDA' for the last month. 2. Calculate the standard deviation of daily percentage returns (volatility). 3. Print the volatility as a percentage. The `agent.run(task)` command executes this prompt.

```python
# The "Pandas" Task
# We ask for a calculation that requires fetching data and doing math.

task = """
Get the historical closing prices for 'NVDA' for the last 1 month.
Calculate the standard deviation of the daily percentage returns (volatility).
Print the volatility as a percentage.
"""

result = agent.run(task, stream=False)
```

```python
print(result)
```

#Task 2 - The Visualization (Plotting)
This is the "Wow" factor. The agent will write code to generate a chart and display it.

The objective here is to showcase the agent's capability to write code for generating and displaying charts using `matplotlib`.

This Python code block defines and executes the second task. The agent is asked to: 1. Plot the closing price of Bitcoin (BTC-USD) for the last 3 months. 2. Add a 7-day moving average line. 3. Save the chart as 'btc_chart.png' and display it. A helper block is included to ensure the image is displayed in Colab if the agent doesn't automatically do so.

```python
# The "Visual" Task
# The agent will use matplotlib to create a chart.

task_viz = """
Plot the closing price of Bitcoin (BTC-USD) for the last 3 months.
Add a 7-day moving average line to the chart.
Save the chart as 'btc_chart.png' and display it.
"""

result = agent.run(task_viz,stream=False)
```

```python
print(result)

# Helper to show the image in Colab if the agent doesn't auto-display
import IPython
if os.path.exists("btc_chart.png"):
    IPython.display.display(IPython.display.Image("btc_chart.png"))
```
